```python

from statistics import *
#Libreria para generar numeros aleatorios
import random
#Libreria de phyton para graficar valores
#as para asignar un alias
import matplotlib.pyplot as plt

#Lista de calificaciones del grupo a de probabilidad
grupoa=[70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print(grupoa)
#La funciÃ³n nos devuelve el tipo de dato de la variable
type(grupoa)
#Sacar la media de las calificaciones con una formula
promedio=sum(grupoa)/len(grupoa)
print(promedio)
#Utilizar la funcion mean que esta en la libreria statistics para calcular el promedio
promedio2=mean(grupoa)
print(promedio2)
#Determinar el valor de la mediana con la funcion median
mediana=median(grupoa)
print(mediana)
#Determinar el valor de la moda
moda=mode(grupoa)
print(moda)
#En caso de que exista mÃ¡s de una moda
#MÃºltiples modas que puedan existir
moda2=multimode(grupoa)
print(moda2)

#Tiempos wue tardaron losalumnos
tiempos100mts=[15.10, 17.20, 14.69, 13.27, 22.15, 18.71, 19.15, 20.3, 15.10, 17.20]
print(tiempos100mts)
#Utilizamos la funcion mean para determinar el tiempo promedio
print("El tiempo promedio de los corredores es =>",mean(tiempos100mts))
#Definimos una variabl3 llamada mediana para calcular la mediana con la funciÃ´n median
mediana=median(tiempos100mts)
print("La media de los tiempos delos corredores es=>", mediana)
#definimos una variable llamada moda para calcular la moda con la funciÃ³n mode
moda=mode(tiempos100mts)
print("La moda de pos tiempos es=>", moda)
multimodas=multimode(tiempos100mts)
print("La moda de los tiempos es=>", multimodas)

fabrica1=[]
for i in range(100):
  numero_aleatorio=round(random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
#Plotter
plt.hist(fabrica1)
plt.title("Produccion diaria de la fabrica 1")
fabrica2=[]
for i in range(100):
  numero_aleatorio2=round(random.uniform(40,93),2)
  fabrica2.append(numero_aleatorio2)
print(fabrica2)




media1=mean(fabrica1)
mediana1=median(fabrica1)
moda1=mode(fabrica1)
print("Valores de la fabrica 1")
print("La media es: ", media1)
print("a mediana es: ", mediana1)
print("La moda es: ", moda1)



```

```python
plt.hist(fabrica2)
plt.title("Produccion diaria de la fabrica 2")
media2=mean(fabrica2)
mediana2=median(fabrica2)
moda2=mode(fabrica2)
print("Valores de la fabrica 2")
print("La media es: ", media2)
print("a mediana es: ", mediana2)
print("La moda es: ", moda2)
```

```python
#Generar 2 subplots de 1 renglon por 2 columnas para desplegar los histogramas, una al lado de la otra
fig, (ax1, ax2)=plt.subplots(nrows=1, ncols=2, figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

```python
plt.hist(fabrica1, bins=10, alpha=0.2, color='b', edgecolor="black")
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label='Mediana')
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label='Moda')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Producion diaria de la fabrica 1')
#Mostrar caja con leyendas
plt.legend()
plt.show()

```
