```python
import math as mt #Para utilizar factorial
from itertools import permutations
from itertools import product
import numpy as np
```

```python
asistentes=["John", "Samantha", "Lucas", "Christian", "Maria", "Helena", "Jacob", "Bryan", "Edgar", "Isabella"]
n=len(asistentes)
r=4
pr=mt.factorial(n)/(mt.factorial(n-r))
print("Hay ",pr," maneras posibles de elegir a los asistentes para calificar el primer examen, habiendo",n,"asistentes y teniendo ",r," preguntas.")
print("Aqui las permutaciones en grupos de ",r," en ",r,".")
permutaciones_asistentes=list(permutations(asistentes, r))
for permutacion in permutaciones_asistentes:
  print(permutacion)
```
