```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
Nombre=["A", "B", "C", "D"]
x=[0, 10, 20, 50]
p=[0.50, 0.30, 0.15, 0.05]
#Verificar que las probabilidades sumen uno
print("Suma de probabilidades", sum(p))
```

```python
#Elaborar la distribucion de probabilidad de x
plt.stem(Nombre,p)
plt.title("Distribucion de probabilidad")
plt.xlabel("Boleto")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```

```python
#Calcular el valor esperado E(X) (ganancia promedio esperada).
valor_esperado=0
#Recorrer sus valores y sus probabilidades
for i in range(len(x)):
  xi=x[i] #Valor de la variable aleatoria
  pi=p[i] #Probabilidad de ese valor
  producto=xi*pi #Multiplicar valor por su probabilidad
  valor_esperado+=producto #Sumar al acumulador
  #Mostrar cada paso
  print(f"x={xi}, p={pi}, x*p={producto:.2f}")
#Mostrar resultado final
print(f"\nValor esperado (Esperanza): {valor_esperado:.2f}")
```

```python
esperanza_cuadrado=0
for i in range(len(x)):
  esperanza_cuadrado+=(x[i]**2)*p[i]
  #Calcular la varianza y la desviacion estandar
  varianza=esperanza_cuadrado-valor_esperado**2
desviación=math.sqrt(varianza)
print(f"E(X)= {valor_esperado:.2f}")
print(f"E(X^2)= {esperanza_cuadrado:.2f}")
print(f"Varianza= {varianza:.2f}")
print(f"Desviación estándar= {desviación:.2f}")
```

```python
#Construir la funcion de distribucion acumulada F(x) y representarla graficamente
F=np.cumsum(p)
#Mostrar resultados
for xi, fi in zip(x, F):
  print(F"f({xi})={fi:.2f}")
#Graficar
plt.step(Nombre, F, where='post')
plt.title("Funcion de Distribucion Acumulada (FDA)")
plt.xlabel("Boleto")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.ylim(0, 1.05)
plt.show()
```
