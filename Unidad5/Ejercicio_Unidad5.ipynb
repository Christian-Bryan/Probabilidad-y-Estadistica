Carga y exploracion de datos

```python
import pandas as pd
import matplotlib.pyplot as plt
#Cargando el primer archivo como DataFrame
df1=pd.read_csv("grupo1.csv")
df1.head()
```

```python
#Cargando el segundo archivo como DataFrame
df2=pd.read_csv("grupo2.csv")
df2.head()
```

```python
#Cargando el tercer archivo como DataFrame
df3=pd.read_csv("grupo3.csv")
df3.head()
```

```python
#Generando histograma de calificaciones del grupo 1
from matplotlib import pyplot
media1=df1.Calificacion.mean()
plt.hist(df1.Calificacion, bins=7)
plt.title("Calificaciones del grupo 1")
plt.axvline(media1, color='y', linestyle='dashed', linewidth=2, label=f'Media={media1:.2f}')
plt.xlabel("Calificaciones")
```

```python
#Generando histograma de calificaciones del grupo 2
media2=df2.Calificacion.mean()
plt.hist(df2.Calificacion, bins=7)
plt.title("Calificaciones del grupo 2")
plt.axvline(media2, color='y', linestyle='dashed', linewidth=2, label=f'Media={media2:.2f}')
plt.xlabel("Calificaciones")
```

```python
#Generando histograma de calificaciones del grupo 3
from matplotlib import pyplot
media3=df3.Calificacion.mean()
plt.hist(df3.Calificacion, bins=7)
plt.title("Calificaciones del grupo 3")
plt.axvline(media3, color='y', linestyle='dashed', linewidth=2, label=f'Media={media3:.2f}')
plt.xlabel("Calificaciones")
```

Modelo de regresion lineal

```python
#Entrenando un modelo de regresion lineal simple
from sklearn.linear_model import LinearRegression
import numpy as np
inversion_tiempo1=np.array(df1.HorasEstudio.tolist()).reshape(-1, 1)
calificacion1=df1.Calificacion.tolist()
modelo1=LinearRegression()
modelo1.fit(inversion_tiempo1, calificacion1)

#Obteniendo la ecuacion de la linea de regresion (pendiente e intercepto)
print("Coeficiente de la pendiente (m):", modelo1.coef_[0])
print("Intercepcion de la recta (b):", modelo1.intercept_)
```

```python
#Entrenando un modelo de regresion lineal simple
from sklearn.linear_model import LinearRegression
import numpy as np
inversion_tiempo2=np.array(df2.HorasEstudio.tolist()).reshape(-1, 1)
calificacion2=df2.Calificacion.tolist()
modelo2=LinearRegression()
modelo2.fit(inversion_tiempo2, calificacion2)

#Obteniendo la ecuacion de la linea de regresion (pendiente e intercepto)
print("Coeficiente de la pendiente (m):", modelo2.coef_[0])
print("Intercepcion de la recta (b):", modelo2.intercept_)
```

```python
#Entrenando un modelo de regresion lineal simple
from sklearn.linear_model import LinearRegression
import numpy as np
inversion_tiempo3=np.array(df3.HorasEstudio.tolist()).reshape(-1, 1)
calificacion3=df3.Calificacion.tolist()
modelo3=LinearRegression()
modelo3.fit(inversion_tiempo3, calificacion3)

#Obteniendo la ecuacion de la linea de regresion (pendiente e intercepto)
print("Coeficiente de la pendiente (m):", modelo3.coef_[0])
print("Intercepcion de la recta (b):", modelo3.intercept_)
```

Calculo de la correlacion

```python
#Calculando el coeficiente de correlacion de Pearson entre las horas de estudio y las calificaciones
from scipy.stats import pearsonr
coeficiente_correlacion, p_valor=pearsonr(df1.HorasEstudio.tolist(), calificacion1)
print("Coeficiente de correlacion de Pearson:", coeficiente_correlacion)
print("Valor p:", p_valor)
```

Visualizacion de resultados

```python
import seaborn as sns

#Incluyendo un grafico de dispersion Scatterplot
plt.figure(figsize=(6, 4))
inversion_tiempo=df1.HorasEstudio.tolist()
sns.set(style="whitegrid")
sns.regplot(x=inversion_tiempo, y=calificacion1, scatter=True, color='blue')
plt.title('Diagrama de dispersion con linea de regresion del grupo 1')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.grid(True)
plt.show()
```
